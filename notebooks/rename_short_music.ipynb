```python
import re
import os
import pandas as pd
```

```python
data = pd.read_csv('/home/amirbek.djanibekov/Documents/nlp703/project/nlp703-speech-processing/csv/usa_music_genre_wo_text.csv')
data.info()
```

```python
data['name_of_the_link'] = data['onmouseover'].apply(
    lambda x: (
            re.search(r'playx\("(?P<name_of_the_link>.*mp3)",', x)
    ).group('name_of_the_link')
)
data['name_of_the_link']
```

```python
files_ = os.listdir('music')
files = dict(
    zip(
        files_,
        range(len(files_))
    )
)
files
```

```python
data['nominal_order'] = data['name_of_the_link'].apply(
    lambda x: files.get(x, 'NotFound404')
)
data['nominal_order']
```

```python
clean_data = data[data['nominal_order'] != 'NotFound404']
clean_data
```

```python
def rename_music_files(row):
    os.rename(row['name_of_the_link'], row['id'] + '.mp3')

clean_data.apply(rename_music_files, axis=1)
```
