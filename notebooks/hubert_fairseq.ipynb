```python
import fairseq
import pandas as pd
import librosa as lr
import os
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
```

```python
path = 'music-segments/'
```

```python
files = os.listdir(path)
df = pd.DataFrame()

for i, file in tqdm(enumerate(files[:10])):
    audio = lr.load(path + file)[0]
    df = pd.concat([df, pd.DataFrame(dict(zip(range(len(audio)), audio)), index=[i])], axis=0)
```

```python
def process_file(file):
    audio = lr.load(path + file)[0]
    return pd.DataFrame(dict(zip(range(len(audio)), audio)), index=[0])

path = 'music-segments/'
files = os.listdir(path)
df = pd.DataFrame()

# Define the number of CPU threads you want to use
num_threads = 8

# Wrap the loop with ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=num_threads) as executor:
    # Process the files in parallel using the executor
    results = list(tqdm(executor.map(process_file, files[:10]), total=len(files[:10])))

# Concatenate the results into a single DataFrame
df = pd.concat(results, axis=0, ignore_index=True)
```
